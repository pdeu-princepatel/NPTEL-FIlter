```python
import pandas as pd

df = pd.read_excel("Final Course List (July - Dec 2025).xlsx")
df = df.drop(range(0,9))
df = df.reset_index()
df = df.drop("index",axis = 1)
df.columns = df.iloc[0]
df.drop(df.index[0], inplace = True)
df.reset_index(drop=True, inplace=True)
```

```python
df 
```

```python
Name = "Computer Science and Engineering"
Duration = "12"
Degr="UG"
# change only above three values to get proper output filtered file
ans = df[df['Discipline'].str.contains(Name, na=False)]
ans = ans[ans["Duration"].str.contains(Duration,na =False)]
ans = ans[ans["UG/PG"].str.contains(Degr,na=False)]
```

```python
ans
```

```python
ans.to_excel("output.xlsx",sheet_name="Required Sheet",index=False)
```

```python
ans.to_csv("output.csv",index=False)
```
